```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
csv_file = '/home/ic40844/myenv/Bias_IndianLanguages/Bias_IndianLanguages/IndiBias_Hindi/Hindi_Results.csv'  # Replace with your actual file path
data = pd.read_csv(csv_file)

# Convert string tuples to actual tuples and split into separate columns
def parse_tuple_column(column):
    return column.apply(lambda x: ast.literal_eval(x))

# Parse each experiment column
for exp in ['Exp 1', 'Exp 2', 'Exp 3', 'Exp 4', 'Exp 5']:
    data[exp] = parse_tuple_column(data[exp])

# Extract stereo and anti-stereo probabilities for each experiment
experiment_data = {}
for exp in ['Exp 1', 'Exp 2', 'Exp 3', 'Exp 4', 'Exp 5']:
    stereo_probs = data[exp].apply(lambda x: x)  # Extract stereo probabilities
    anti_probs = 1- stereo_probs    # Extract anti-stereo probabilities
    experiment_data[exp] = {'stereo': stereo_probs.tolist(), 'anti': anti_probs.tolist()}

# Prepare data for plotting
experiments = ['Exp 1', 'Exp 2', 'Exp 3', 'Exp 4', 'Exp 5']
stereo_means = [np.mean(experiment_data[exp]['stereo']) for exp in experiments]
anti_means = [np.mean(experiment_data[exp]['anti']) for exp in experiments]

# Create a grouped bar plot
x = np.arange(len(experiments))  # X-axis positions for experiments
width = 0.35                     # Width of the bars

fig, ax = plt.subplots(figsize=(10, 6))

# Plot stereo and anti-stereo bars
ax.bar(x - width/2, stereo_means, width, label='Stereotypical', color='#d73027')
ax.bar(x + width/2, anti_means, width, label='Anti-Stereotypical', color='#4575b4')

# Add labels, title, and legend
ax.set_xlabel('Experiments')
ax.set_ylabel('Average Probability (%)')
ax.set_title('Stereo vs Anti-Stereo Probabilities Across Experiments')
ax.set_xticks(x)
ax.set_xticklabels(experiments)
ax.legend()

# Annotate values on bars
for i in range(len(experiments)):
    ax.text(x[i] - width/2, stereo_means[i] + 0.5, f'{stereo_means[i]:.1f}%', ha='center', va='bottom', color='black')
    ax.text(x[i] + width/2, anti_means[i] + 0.5, f'{anti_means[i]:.1f}%', ha='center', va='bottom', color='black')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the CSV file
csv_file = '/home/ic40844/myenv/Bias_IndianLanguages/Blank_experiments/IndiBias_Hindi_Results/Blank_Hindi_Results.csv'  # Replace with your actual file path
data = pd.read_csv(csv_file)
data= data.head(1)
# Extract stereo probabilities for each experiment
experiments = [col for col in data.columns if 'Exp' in col]  # Identify experiment columns

# Prepare data for plotting
stereo_data = {exp: data[exp].astype(float).tolist() for exp in experiments}  # Convert to float

# Create a horizontal bar chart
fig, ax = plt.subplots(figsize=(10, 8))
bar_width = 0.8 / len(experiments)  # Adjust bar width based on number of experiments

# Plot each experiment's stereo probabilities as bars
for i, (exp, probs) in enumerate(stereo_data.items()):
    y_positions = range(len(probs))
    ax.barh([y + i * bar_width for y in y_positions], probs, bar_width, label=exp)

# Add labels to each bar
for i, (exp, probs) in enumerate(stereo_data.items()):
    for j, prob in enumerate(probs):
        ax.text(prob + 0.02, j + i * bar_width, f'{exp}', ha='left', va='center', fontsize=8)

# Formatting
ax.set_yticks([y + (len(experiments) - 1) * bar_width / 2 for y in range(len(data))])  # Center labels for rows
# ax.set_yticklabels([f'Row {i+1}' for i in range(len(data))])
ax.set_xlabel('Stereo Probability')
ax.set_ylabel('Experiment with Hindi (Airavata) on IndiBias')
ax.set_xlim(0, 1)  # Assuming probabilities are between 0 and 1
ax.set_title('Stereo Probabilities Across Experiments')
ax.legend(loc='upper right', title='Experiments')
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Remove extra tick marks on the Y-axis
ax.tick_params(axis='y', left=False)

plt.tight_layout()
plt.show()
```
